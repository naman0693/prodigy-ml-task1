```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
from google.colab import drive

# Mount Google Drive to access files
google_drive_mount_path = '/content/drive'
drive.mount(google_drive_mount_path)
```

```python
# Load the training dataset
housing_train_data = pd.read_csv('train.csv')

# Display the first few rows of the training data
housing_train_data.head()
```

```python
# Select relevant features and target variable from the training data
housing_features = housing_train_data[['GrLivArea', 'BedroomAbvGr', 'SalePrice']].copy()

# Remove any rows with missing values
housing_features.dropna(inplace=True)
```

```python
# Split into features (predictor variables) and target (what we want to predict)
features = housing_features[['GrLivArea', 'BedroomAbvGr']]  # Living area and bedroom count
target = housing_features['SalePrice']  # Housing price we want to predict

# Split into training and test sets (80% train, 20% test)
features_train, features_test, target_train, target_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)
```

```python
# Initialize the linear regression model
price_predictor = LinearRegression()

# Train the model on our training data
price_predictor.fit(features_train, target_train)
```

```python
# Make predictions on the test set
test_predictions = price_predictor.predict(features_test)

# Calculate evaluation metrics
test_mean_squared_error = mean_squared_error(target_test, test_predictions)
test_r2_score = r2_score(target_test, test_predictions)

# Display model performance metrics
print("\nModel Performance Evaluation:")
print(f"Mean Squared Error (MSE): {test_mean_squared_error:,.2f}")
print(f"R² Score (Coefficient of Determination): {test_r2_score:.2%}")

# Optional: Create a results DataFrame
performance_metrics = pd.DataFrame({
    'Metric': ['Mean Squared Error', 'R² Score'],
    'Value': [test_mean_squared_error, test_r2_score]
})
```

```python
# Load the test dataset
test_data = pd.read_csv('test.csv')

# Select the same features used for training
final_features = test_data[['GrLivArea', 'BedroomAbvGr']]

# Generate predictions using our trained model
final_predictions = price_predictor.predict(final_features)

# Prepare submission file with IDs and predicted prices
submission_data = pd.DataFrame({
    'Id': test_data['Id'],
    'SalePrice': final_predictions
})

# Save predictions to CSV file without index
submission_data.to_csv('house_price_predictions.csv', index=False)

# Confirmation message
print("Submission file 'house_price_predictions.csv' successfully created.")
print(f"Number of predictions made: {len(final_predictions)}")
```

```python
# Create a figure with custom size for better readability
plt.figure(figsize=(10, 7))

# Generate the scatter plot with semi-transparent points
plt.scatter(x=target_test, y=test_predictions,
            alpha=0.6, color='royalblue', edgecolor='w')

# Add perfect prediction line (45-degree line)
perfect_line = [target_test.min(), target_test.max()]
plt.plot(perfect_line, perfect_line,
         color='red', linestyle='--', linewidth=1,
         label='Perfect Prediction')

# Add labels and title with custom font sizes
plt.xlabel('Actual Sale Price ($)', fontsize=12, labelpad=10)
plt.ylabel('Predicted Sale Price ($)', fontsize=12, labelpad=10)
plt.title('Actual vs Predicted Home Prices\nModel Performance Evaluation',
          fontsize=14, pad=15)

# Add grid and adjust layout
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Add legend and customize ticks
plt.legend(fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Import the ticker module for formatting axis labels
import matplotlib.ticker as ticker

# Format axis to use dollar signs for prices
formatter = ticker.StrMethodFormatter('${x:,.0f}')
plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)

# Display the plot
plt.show()
```
